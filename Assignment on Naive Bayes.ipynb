## Label encoding on both data

```python
le = LabelEncoder()
training_data['Salary'] = le.fit_transform(training_data['Salary'])
test_data['Salary'] = le.fit_transform(test_data['Salary'])
```

```python
training_data.head()
```

```python
test_data.head()
```

```python
#  Label Encoding on Training data
training_data = pd.get_dummies(data= training_data, columns= ['workclass','education','maritalstatus','occupation','relationship','race','sex','native'])
```

```python
training_data
```

```python
#  Label Encoding on Testing data
test_data = pd.get_dummies(data= test_data, columns= ['workclass','education','maritalstatus','occupation','relationship','race','sex','native'])
```

```python
test_data
```

###  Splitting data into input and output variables

```python
# On training data
x_train = training_data.drop(labels= 'Salary', axis = 1)
y_train = training_data[['Salary']]
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# On testing data
x_test = test_data.drop(labels='Salary',axis = 1)
y_test = test_data[['Salary']]
```

```python
x_test.head()
```

```python
y_test.head()
```

# Building model

## Gaussian Naive Bayes

```python
# On training data
train_model = gb()
train_model.fit(x_train,y_train)
```

```python
y_train_pred = train_model.predict(x_train)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
# On testing data
y_test_pred = train_model.predict(x_test)
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
print('Training data accurancy : ',accuracy_score(y_train,y_train_pred))
print('Testing data accurancy : ',accuracy_score(y_test,y_test_pred))
```

## Multinominal Naive Bayes

```python
mb_model = mb()
mb_model.fit(x_train,y_train)
```

```python
y_mb_train = mb_model.predict(x_train)
```

```python
# On training data
accuracy_score(y_train,y_mb_train)
```

```python
print(classification_report(y_train,y_mb_train))
```

```python
# On testing data
y_test_mb = mb_model.predict(x_test)
```

```python
accuracy_score(y_test,y_test_mb)
```

```python
print(classification_report(y_test,y_test_mb))
```

```python
print('Training data accurancy : ',accuracy_score(y_train,y_mb_train))
print('Testing data accurancy : ',accuracy_score(y_test,y_test_mb))
```
